```python
import sys
sys.path.append('..')
import numpy as np
import pandas as pd
from pupeyes.saccades import saccade_aoi_angular, saccade_deviation
import pickle

%load_ext autoreload
%autoreload 2
```

# Saccade AOI Angular

```python
data, sample_data = pickle.load(open('saccade_angular_tests.pkl', 'rb'))
```

```python
item_coords = [[0,300], [300,0], [0,-300], [-300,0]]
```

```python
# compute saccade curvature and destination
# Get unique participant values
unique_participants = data['participant'].unique()

# Initialize an empty list to store computed saccade curvatures
full_saccade_destination_list = []

# Loop through unique participants for faster filtering
for participant in unique_participants:
    # Filter raw_eyelink_samples for the current participant
    participant_samples = sample_data[sample_data['participant'] == participant]
    participant_data = data[data['participant'] == participant]

    # computes saccade destination based on full saccade trajectory
    df = saccade_aoi_angular(sample_data = participant_samples,
                        data=participant_data,
                        col_sample_timestamp='trackertime',
                        col_x='x',
                        col_y='y',
                        col_saccade_start_time='starttime',
                        col_saccade_end_time='endtime',
                        col_target_pos='target_pos',
                        col_distractor_pos='distractor_pos',
                        col_distractor_cond='distractor_cond',
                        col_other_pos=None,
                        item_coords=item_coords,
                        use = None,
                        threshold=30)
    full_saccade_destination_list.append(df)

data = pd.concat(full_saccade_destination_list)
```

```python
data.full_saccade_destination.equals(data.curritem)
```

# Saccade Curvature


```python
# compute saccade curvature and destination
# Get unique participant values
unique_participants = data['participant'].unique()

# Initialize an empty list to store computed saccade curvatures
saccade_curvature_list = []

# Loop through unique participants for faster filtering
for participant in unique_participants:
    # Filter raw_eyelink_samples for the current participant
    participant_samples = sample_data[sample_data['participant'] == participant]
    participant_data = data[data['participant'] == participant]

    # computes saccade destination based on full saccade trajectory
    df = saccade_deviation(
        sample_data = participant_samples, 
        data = participant_data,
        col_sample_timestamp='trackertime',
        col_saccade_start_time='starttime',
        col_saccade_end_time='endtime',
        col_x='x',
        col_y='y',
        find = 'mid'
        )
    saccade_curvature_list.append(df)

curvature_data = pd.concat(saccade_curvature_list)
```
